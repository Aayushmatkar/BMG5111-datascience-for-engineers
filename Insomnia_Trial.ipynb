```python
# prompt: code to train a model to predict insomnia using the sleepdata.csv

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Read a CSV file into a Pandas DataFrame
df = pd.read_csv('sleepdata.csv')

# Separate features and target
X = df[['Quality of Sleep', 'Sleep Duration', 'Stress Level']]
y = df['Sleep Disorder']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Train a logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Evaluate the model on the testing set
score = model.score(X_test, y_test)
print(f"Model accuracy: {score}")

```

```python
# prompt: write a code to predict an insomnia affected persoon

# Get user input for features
quality_of_sleep = float(input("Enter your quality of sleep (1-10): "))
sleep_duration = float(input("Enter your sleep duration (hours): "))
stress_level = float(input("Enter your stress level (1-10): "))

# Create a feature vector
features = [quality_of_sleep, sleep_duration, stress_level]

# Predict the probability of insomnia
predicted_probability = model.predict_proba([features])[0][1]
print(predicted_probability)
# Print the prediction
if predicted_probability > 21:
  print("You are likely to have insomnia.")
else:
  print("You are unlikely to have insomnia.")
```

```python
# prompt: write a code to test the model

# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# Read a CSV file into a Pandas DataFrame
df = pd.read_csv('sleepdata.csv')

# Separate features and target
X = df[['Quality of Sleep', 'Sleep Duration', 'Stress Level']]
y = df['Sleep Disorder']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

# Train a logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Evaluate the model on the testing set
score = model.score(X_test, y_test)
print(f"Model accuracy: {score}")

# Get user input for features
quality_of_sleep = float(input("Enter your quality of sleep (1-10): "))
sleep_duration = float(input("Enter your sleep duration (hours): "))
stress_level = float(input("Enter your stress level (1-10): "))

# Create a feature vector
features = [quality_of_sleep, sleep_duration, stress_level]

# Predict the probability of insomnia
predicted_probability = model.predict_proba([features])[0][1]
print(predicted_probability)



# Test the model with new data
new_data = [[5, 7, 8], [3, 5, 4], [9, 10, 10]]
predicted_labels = model.predict(new_data)

# Print the predictions for new data
for i, label in enumerate(predicted_labels):
  print(f"Data point {i+1}: {new_data[i]}, Predicted label: {label}")
```
